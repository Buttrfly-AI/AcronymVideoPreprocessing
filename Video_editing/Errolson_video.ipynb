```python
# lets get three different videos
# part 1
!ffmpeg -i   video_pipeline.mp4 -ss 00:00:00 -t 00:00:05 -async 1 part_1_whole.mp4 -y
```

```python
# part 2
!ffmpeg -i video_pipeline.mp4 -ss 00:00:05 -to 00:00:10 -async 1 part_2_whole.mp4 -y
!ffmpeg -i part_2_whole.mp4 -vf "rotate=PI:bilinear=0,format=yuv420p" -metadata:s:v rotate=0 -codec:v libx264 -codec:a -async 1 part_2_rotated.mp4
!ffmpeg -i part_2_rotated.mp4 -vf hue=s=0  -async 1 part_2_final.mp4
```

```python
# part 3
!ffmpeg -i video_pipeline.mp4 -ss 00:00:10 -to 00:00:15 -async 1  part_3_whole.mp4 -y


!ffmpeg -i part_3_whole.mp4 -vf colorbalance=rs=.3  -async 1 part_3_final.mp4

```
